```python
import yfinance as yf
```

```python
# testing ollama model to give ticker name
from langchain_ollama import ChatOllama
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun


@tool
def stock_info(ticker_name: str) -> dict:
    """Find information about ticker by providing ticker symbol

    Args:
        ticker_name (str): ticker symbol of the company

    Returns:
        dict: details about the company
    """
    ticker = yf.Ticker(ticker_name)
    return {f"{ticker_name}_info": ticker.info}

@tool
def internet_search(query: str):
    """Run a web search
    Args:
        query (str): The query presented by user to search on internet.
    """
    search = DuckDuckGoSearchRun()
    return search.run(query)
```

```python
# static agent
import os
from langchain.agents import create_agent

static_stock_agent = create_agent(
    model=ChatOllama(model='qwen3:8b'),
    tools=[internet_search, stock_info],
    system_prompt="You are an agent for finding news related information of any query."
)
```

```python
response = static_stock_agent.invoke({
    "input": "find news on Russia war"})
```
